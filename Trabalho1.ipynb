# 1) Gerador Linear Congruente (GLC)

Considere o gerador linear congruente definido por:

 X_{n+1} = (aX_n + c) mod m

com parâmetros:
- a = 11
- c = 7
- m = 32
- X0 = 5

      a) calcular os 10 primeiros números gerados;
      b) determinar o período do gerador;
      c) verificar se o gerador satisfaz o período máximo;
      d) discutir limitações para simulações de Monte Carlo em larga escala.


```python
# Parâmetros do gerador
a = 11
c = 7
m = 32
x0 = 5
```

```python
# --------------------------------------------
# a) Calcular os 10 primeiros números gerados
# --------------------------------------------
x=x0
# Lista onde serão armazenados os valores gerados
primeiros_10 = []
# Gerar 10 valores usando a fórmula do GLC
for i in range(10):
    x = (a * x+ c) % m   # aplica X_{n+1} = (aX_n + c) mod m
    primeiros_10.append(x)

# Mostrar o resultado
print("a) Os 10 primeiros números gerados são:")
print(primeiros_10)
```

```python
# --------------------------------------------
# b) Determinar o período do gerador
# --------------------------------------------
x=x0
# Guardar o valor inicial
x_inicial = x0
# Lista para armazenar a sequência completa
sequencia = [x_inicial]
# Contador do período
periodo = 0
# Gerar números até a semente inicial aparecer novamente
while True:
    x = (a * x + c) % m   # gera o próximo valor
    sequencia.append(x)   # adiciona o valor à sequência
    periodo += 1          # incrementa o contador

    # Se voltou ao valor inicial, o ciclo se completou
    if x == x_inicial:
        break

# Mostrar a sequência até repetir
print("\nb) Sequência até repetir a semente inicial:")
print(sequencia)

# Mostrar o período encontrado
print("Período do gerador:", periodo)
```

```python
# --------------------------------------------
# Gráfico da sequência gerada
# --------------------------------------------

# Criar os índices n para o eixo x
indices = list(range(len(sequencia)))

# Criar o gráfico
plt.figure(figsize=(10, 5))
plt.bar(indices, sequencia, color='blue')
plt.title("Sequência gerada pelo Gerador Linear Congruente")
plt.xlabel("Índice n")
plt.ylabel("Valor Xn")
plt.grid(True)
plt.show()
```

# 2) Distribuição de Poisson

Uma central de monitoramento recebe eventos de falha em um sistema IoT de sensores agrícolas.

O número médio de falhas observadas é 4 falhas por hora.

    a) Qual é a probabilidade de ocorrerem exatamente 6 falhas em uma hora?
    b) Qual é a probabilidade de ocorrerem no máximo 2 falhas em uma hora?
    c) Qual é a probabilidade de ocorrerem mais de 5 falhas em duas horas?
    d) Calcule o valor esperado e a variância da variável aleatória associada.

```python
# Média de falhas por hora
lamb = 4
```

```python
# --------------------------------------------
# a) Probabilidade de ocorrerem exatamente 6 falhas em 1 hora
# Fórmula da distribuição de Poisson:
# P(X = k) = (e^(-lambda) * lambda^k) / k!
# --------------------------------------------

# Definir o número de falhas desejado
k = 6

# Aplicar a fórmula da Poisson
prob_a = (math.exp(-lamb) * (lamb ** k)) / math.factorial(k)

# Mostrar o resultado do item a
print("a) Probabilidade de ocorrerem exatamente 6 falhas em 1 hora:")
print(prob_a, "ou", round(prob_a * 100, 2), "%")
```

```python
# --------------------------------------------
# b) Probabilidade de ocorrerem no máximo 2 falhas em 1 hora
# "No máximo 2" significa:
# P(X <= 2) = P(X=0) + P(X=1) + P(X=2)
# --------------------------------------------

# Inicializar a probabilidade acumulada com zero
prob_b = 0

# Somar as probabilidades de 0, 1 e 2 falhas
for k in range(0, 3):
    prob_b += (math.exp(-lamb) * (lamb ** k)) / math.factorial(k)

# Mostrar o resultado do item b
print("\nb) Probabilidade de ocorrerem no máximo 2 falhas em 1 hora:")
print(prob_b, "ou", round(prob_b * 100, 2), "%")
```

```python
# --------------------------------------------
# c) Probabilidade de ocorrerem mais de 5 falhas em 2 horas
# Em 2 horas, a média passa a ser:
# lambda_2h = 4 * 2 = 8
#
# "Mais de 5" significa:
# P(X > 5)
#
# É mais fácil calcular pelo complemento:
# P(X > 5) = 1 - P(X <= 5)
# --------------------------------------------

# Definir a média de falhas em 2 horas
lamb_2h = 8

# Inicializar a soma das probabilidades de 0 até 5 falhas
prob_ate_5 = 0

# Somar as probabilidades de 0, 1, 2, 3, 4 e 5 falhas
for k in range(0, 6):
    prob_ate_5 += (math.exp(-lamb_2h) * (lamb_2h ** k)) / math.factorial(k)

# Calcular a probabilidade complementar
prob_c = 1 - prob_ate_5

# Mostrar o resultado do item c
print("\nc) Probabilidade de ocorrerem mais de 5 falhas em 2 horas:")
print(prob_c, "ou", round(prob_c * 100, 2), "%")
```

```python
# --------------------------------------------
# d) Valor esperado e variância
# Para a distribuição de Poisson:
# Valor esperado = lambda
# Variância = lambda
# Como o enunciado fala da variável por hora, usamos lambda = 4
# --------------------------------------------

# Calcular o valor esperado
valor_esperado = lamb

# Calcular a variância
variancia = lamb

# Mostrar os resultados do item d
print("\nd) Valor esperado e variância da variável aleatória:")
print("Valor esperado =", valor_esperado)
print("Variância =", variancia)
```

# 3) Distribuição Binomial

Em um sistema de comunicação digital, cada pacote transmitido tem probabilidade de erro igual a 0.08.
Considere o envio de 20 pacotes.

    a) Determine a probabilidade de ocorrerem exatamente 3 erros.
    b) Determine a probabilidade de ocorrerem no máximo 2 erros.
    c) Determine a probabilidade de ocorrerem mais de 5 erros.
    d) Calcule média e variância da distribuição.

```python
# Fixar a semente para reprodutibilidade
np.random.seed(42)

# Número de simulações
N = 10000000

# Parâmetros da binomial
n = 20          # número de pacotes
q = 0.08        # probabilidade de erro em cada pacote
p = 1 - q       # probabilidade de não ocorrer erro

# Lista para armazenar as amostras geradas
amostras_bin = []

# --------------------------------------------
# Gerar amostras da Binomial(n,q)
# usando o algoritmo recursivo do PDF
# --------------------------------------------

for _ in range(N):

    # 1) Gerar uma variável uniforme U(0,1)
    u = np.random.rand()

    # 2) Calcular:
    # c = q / (1-q)
    # pr = (1-q)^n = P(X=0)
    c = q / (1 - q)
    pr = (1 - q) ** n

    # 3) Inicializar:
    # i = 0
    # F = pr
    i = 0
    F = pr

    # 4) Se u < F, retorna X = i
    # 5) Caso contrário, atualiza pr recursivamente
    #    e acumula em F até encontrar o intervalo correspondente
    while u >= F:
        pr = c * ((n - i) / (i + 1)) * pr
        F = F + pr
        i = i + 1

    # Guardar a amostra gerada
    amostras_bin.append(i)

# Transformar em array para facilitar os cálculos
amostras_bin = np.array(amostras_bin)
```

```python
# --------------------------------------------
# a) Probabilidade de ocorrerem exatamente 3 erros
# --------------------------------------------
# frequência relativa de X = 3
prob_a_sim = np.mean(amostras_bin == 3)

print("a) Probabilidade de ocorrerem exatamente 3 erros")
print(prob_a_sim, "ou", round(prob_a_sim * 100, 2), "%")
```

```python
# --------------------------------------------
# b) Probabilidade de ocorrerem no máximo 2 erros
# --------------------------------------------

# frequência relativa de X <= 2
prob_b_sim = np.mean(amostras_bin <= 2)

print("\nb) Probabilidade de ocorrerem no máximo 2 erros")
print(prob_b_sim, "ou", round(prob_b_sim * 100, 2), "%")
```

```python
# --------------------------------------------
# c) Probabilidade de ocorrerem mais de 5 erros
# --------------------------------------------

# frequência relativa de X > 5
prob_c_sim = np.mean(amostras_bin > 5)


print("\nc) Probabilidade de ocorrerem mais de 5 erros")
print(prob_c_sim, "ou", round(prob_c_sim * 100, 2), "%")
```

```python
# --------------------------------------------
# d) Média e variância
# --------------------------------------------
media_sim = np.mean(amostras_bin)
variancia_sim = np.var(amostras_bin)

print("\nd) Média e variância")
print("Média:", media_sim)
print("Variância:", variancia_sim)
```

# 4) Distribuição Geométrica

A probabilidade de serem necessários n insucessos antes do primeiro sucesso é

P(Y = n) =((1 - p)^n)p

para n = 0,1,2,3,.......

Um sistema de comunicação repete transmissões até receber confirmação de recebimento. A probabilidade de sucesso em cada tentativa é p = 0.3.

    a) Criar uma simulação para gerar amostras da distribuição geométrica
    b) Qual a probabilidade de o primeiro sucesso ocorrer na 4ª tentativa?
    c) Qual a probabilidade de precisar de mais de 6 tentativas?
    d) Determine a média e variância da variável aleatória.

```python
# --------------------------------------------
# a) Criar uma simulação para gerar amostras
# da distribuição geométrica
# --------------------------------------------

# Fixar a semente para reprodutibilidade
np.random.seed(42)

# Número de simulações
N = 100000

# Probabilidade de sucesso em cada tentativa
p = 0.3

# Lista para armazenar as amostras geradas
amostras_geo = []

# Repetir o experimento N vezes
for _ in range(N):

    # Contador de insucessos antes do primeiro sucesso
    y = 0

    # Repetir tentativas até ocorrer o primeiro sucesso
    while True:

        # Gerar uma variável uniforme entre 0 e 1
        u = np.random.rand()

        # Se u < p, consideramos que ocorreu sucesso
        # e encerramos esta repetição
        if u < p:
            break

        # Caso contrário, ocorreu insucesso
        # e aumentamos o contador
        y += 1

    # Guardar a amostra gerada
    amostras_geo.append(y)

# Transformar em array para facilitar os cálculos
amostras_geo = np.array(amostras_geo)

# Mostrar algumas amostras geradas
print("a) Algumas amostras geradas da distribuição geométrica:")
print(amostras_geo[:20])
```

```python
# --------------------------------------------
# b) Probabilidade de o primeiro sucesso ocorrer na 4ª tentativa
#
# Se o primeiro sucesso ocorre na 4ª tentativa, então houve 3 insucessos antes do primeiro sucesso.
# Logo, isso equivale a Y = 3.
# --------------------------------------------

# frequência relativa de Y = 3
prob_b_sim = np.mean(amostras_geo == 3)

print("\nb) Probabilidade de o primeiro sucesso ocorrer na 4ª tentativa")
print(prob_b_sim, "ou", round(prob_b_sim * 100, 2), "%")
```

```python
# --------------------------------------------
# c) Probabilidade de precisar de mais de 6 tentativas
#
# Precisar de mais de 6 tentativas significa que o sucesso não ocorreu nas 6 primeiras tentativas.
#
# Como Y conta o número de insucessos antes do primeiro sucesso, isso equivale a Y >= 6.
# --------------------------------------------

# frequência relativa de Y >= 6
prob_c_sim = np.mean(amostras_geo >= 6)

print("\nc) Probabilidade de precisar de mais de 6 tentativas")
print(prob_c_sim, "ou", round(prob_c_sim * 100, 2), "%")
```

```python
# --------------------------------------------
# d) Média e variância da distribuição
#
# Para Y = número de insucessos antes do primeiro sucesso:
# média = (1-p)/p
# variância = (1-p)/(p^2)
# --------------------------------------------

# Valores simulados
media_sim = np.mean(amostras_geo)
variancia_sim = np.var(amostras_geo)

print("\nd) Média e variância")
print("Média simulada    :", media_sim)
print("Variância simulada:", variancia_sim)
```

# 5) Método da Inversa

Considere a variável aleatória com pdf:

f(x) = 2x, 0 ≤ x ≤ 1

    a) Verifique que f(x) é uma densidade válida.
    b) Determine a função distribuição acumulada (CDF).
    c)Com uso da transformação inversa, desenvolva um algoritmo para gerar amostras dessa distribuição.
    e) Gere 1000 amostras e compare o histograma com a pdf teórica.

```python
# --------------------------------------------
# c) Usando a transformação inversa
#
# Se U ~ Uniforme(0,1), então:
# U = F(X) = X^2
# X = sqrt(U)
#
# --------------------------------------------
# e) Gerar 1000 amostras e comparar o histograma
# com a pdf teórica
# --------------------------------------------

# Número de amostras
N = 1000

# Gerar um vetor U com distribuição uniforme entre 0 e 1
u = np.random.uniform(0, 1, N)

# Aplicar a transformação inversa
x_amostras = np.sqrt(u)

# Mostrar algumas amostras geradas
#print("\nPrimeiras 1000 amostras geradas:")
#print(x_amostras[:1000])

# Criar valores de x para desenhar a pdf teórica
x_teorico = np.linspace(0, 1, 400)

# Calcular a pdf teórica f(x) = 2x
f_teorica = 2 * x_teorico

# Criar o histograma das amostras
plt.figure(figsize=(8, 5))
plt.hist(x_amostras, bins=25, density=True, alpha=0.7, edgecolor='black', label='Histograma das amostras')

# Desenhar a pdf teórica
plt.plot(x_teorico, f_teorica, linewidth=2, label='pdf teórica: f(x)=2x')

# Configurar o gráfico
plt.title("Método da Inversa: Histograma vs pdf teórica")
plt.xlabel("x")
plt.ylabel("Densidade")
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()
```
